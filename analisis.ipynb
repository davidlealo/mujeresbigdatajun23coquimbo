Análisis de Emisiones a cuerpos de agua

```python
import pandas as pd
```

```python
# Emisiones a cuerpos de agua 2020
emisiones_agua_2020 = pd.read_excel('riles_2020_v3.xlsx')
```

```python
# Ejemplo de shape de un dataframe realizando nuevamente la carga de datos
pd.read_excel('riles_2020_v3.xlsx').shape
```

```python
# Ejemplo de shape de un dataframe cuya carga está asignada a una variable
emisiones_agua_2020.shape
```

```python
emisiones_agua_2020.info()
```

```python
# cantidad de nulos por variable
emisiones_agua_2020.isnull().sum()
```

```python
# Agrupar por vu los nombres de establecimientos

nombre_establecimiento = emisiones_agua_2020.groupby('vu')['Nombre de establecimiento'].unique()
display(nombre_establecimiento)
```

```python
# CIIU4  unicos
clasicficacion_economica_unicos_ciiu4 = emisiones_agua_2020.groupby('CIIU4')['Nombre de establecimiento'].unique()
display(clasicficacion_economica_unicos_ciiu4)

```

```python
# Exportar dataframe
clasicficacion_economica_unicos_ciiu4.to_csv('clasicficacion_economica_unicos_ciiu4.csv')
```

```python
# CIIU6  unicos
clasicficacion_economica_unicos_ciiu6 = emisiones_agua_2020.groupby('CIIU6')['Nombre de establecimiento'].unique()
display(clasicficacion_economica_unicos_ciiu6)
```

```python
clasicficacion_economica_unicos_ciiu6.to_csv('clasicficacion_economica_unicos_ciiu6.csv', index=False)
```

```python
# listado de origen unicos
origen = emisiones_agua_2020['Origen'].unique()
display(origen)
```

```python
# tipo de contaminante por region
tipo_contaminante_region = emisiones_agua_2020.groupby(['Región', 'Tipo de contaminate']).size().reset_index(name='count')
tipo_contaminante_region.to_csv('tipo_contaminante_region.csv', index=False)
```

```python
# grafico de tipo de contaminante en region antofagasta
tipo_contaminante_antofagasta = emisiones_agua_2020[emisiones_agua_2020['Región'] == 'Antofagasta'].groupby('Tipo de contaminate').size().sort_values(ascending=False)
display(tipo_contaminante_antofagasta)
```
